```python
def make_mp_date(s):
    ymd = s.split('-')
    y = ymd[0]
    m = ymd[1]
    d = ymd[2]
    if len(m) ==1:
        m = '0'+m
    if len(d) == 1:
        d = '0'+d
    return int(y+m+d)
```
